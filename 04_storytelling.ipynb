# Análisis del Mercado del Cobre en Chile 1990-2024

## Contexto
Chile es el mayor productor de cobre del mundo, representando 
aproximadamente el 27% de la producción global. El precio del 
cobre es uno de los indicadores económicos más importantes para 
Chile, afectando directamente los ingresos fiscales, el tipo de 
cambio y el crecimiento del PIB.

## Hallazgos principales

### 1. El cobre multiplicó su valor 4 veces en 30 años
El precio promedio pasó de 101 cUS$/lb en los años 90 a 
380 cUS$/lb en la década 2020-2024, impulsado por la 
industrialización de China y la transición energética global.

### 2. Los ciclos del cobre siguen los ciclos globales
- **1998:** Crisis asiática → caída de -27.4%
- **2008:** Crisis financiera global → caída de -25.7%
- **2021:** Boom post-pandemia → alza de +50.7% (peak histórico)

### 3. Sin estacionalidad mensual significativa
A diferencia de commodities agrícolas, el cobre no muestra 
patrones estacionales fuertes. Sus precios responden a factores 
macroeconómicos globales como la demanda china, el dólar y 
las tasas de interés.

## Conclusión
El cobre chileno está en su mejor momento histórico de precios, 
impulsado por la demanda de la transición energética (vehículos 
eléctricos, energías renovables). Chile debe aprovechar este 
ciclo para diversificar su economía y agregar valor a la cadena 
minera.
