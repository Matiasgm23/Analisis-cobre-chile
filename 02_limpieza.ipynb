```python
import pandas as pd
import numpy as np

# Cargar ambos datasets limpios
df_anual = pd.read_excel('data/Precios-del-Cobre-Refinado-Anual.xlsx',
                          skiprows=4,
                          names=['Año', 'Precio_BML', 'Precio_COMEX'])
df_anual = df_anual.dropna(subset=['Año'])
df_anual = df_anual[df_anual['Año'] != 'AÑO']
df_anual['Año'] = pd.to_numeric(df_anual['Año'], errors='coerce')
df_anual['Precio_BML'] = pd.to_numeric(df_anual['Precio_BML'], errors='coerce')
df_anual = df_anual.dropna(subset=['Año', 'Precio_BML'])
df_anual['Año'] = df_anual['Año'].astype(int)

df_mensual = pd.read_excel('data/Precios-del-Cobre-Refinado-Mensual.xlsx',
                            skiprows=4,
                            names=['Fecha', 'Precio_BML', 'Precio_COMEX'])
df_mensual = df_mensual.dropna(subset=['Fecha'])
df_mensual = df_mensual[df_mensual['Fecha'] != 'MES - AÑO']
df_mensual['Fecha'] = pd.to_datetime(df_mensual['Fecha'], errors='coerce')
df_mensual['Precio_BML'] = pd.to_numeric(df_mensual['Precio_BML'], errors='coerce')
df_mensual = df_mensual.dropna(subset=['Fecha', 'Precio_BML'])
df_mensual['Año'] = df_mensual['Fecha'].dt.year
df_mensual['Mes'] = df_mensual['Fecha'].dt.month

print("Anual:", df_anual.shape)
print("Mensual:", df_mensual.shape)
```

```python
# Columnas derivadas para dataset anual
df_anual['Variacion_Anual_%'] = df_anual['Precio_BML'].pct_change() * 100
df_anual['Promedio_Movil_5'] = df_anual['Precio_BML'].rolling(window=5).mean()
df_anual['Década'] = (df_anual['Año'] // 10) * 10

# Columnas derivadas para dataset mensual
df_mensual['Variacion_Mensual_%'] = df_mensual['Precio_BML'].pct_change() * 100
df_mensual['Promedio_Movil_12'] = df_mensual['Precio_BML'].rolling(window=12).mean()

# Filtrar desde 1990 para análisis más relevante
df_anual_mod = df_anual[df_anual['Año'] >= 1990].copy()
df_mensual_mod = df_mensual[df_mensual['Año'] >= 1990].copy()

print("Anual desde 1990:", df_anual_mod.shape)
print("Mensual desde 1990:", df_mensual_mod.shape)
print("\nEjemplo anual:")
print(df_anual_mod.tail(5))
```

```python
# Guardar datasets limpios
df_anual_mod.to_csv('data/cobre_anual_limpio.csv', index=False, sep=';', decimal=',')
df_mensual_mod.to_csv('data/cobre_mensual_limpio.csv', index=False, sep=';', decimal=',')

print("Archivos guardados correctamente")
print("\nResumen anual:")
print(df_anual_mod[['Año', 'Precio_BML', 'Variacion_Anual_%']].describe().round(2))
```
